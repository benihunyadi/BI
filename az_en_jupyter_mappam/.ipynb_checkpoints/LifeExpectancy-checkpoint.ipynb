```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
dataset = pd.read_csv('LifeExpectancy.csv')
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
#Ellenőrizi, hogy mely oszlopok tartalmaznak  NaN értékeket 
```

```python
dataset.isnull().any()
```

```python
#Ha bármely oszlopban true eredményt talál, akkor az alábbi kód használatával eltavolitja  az összes oszlopbol a null értékét.
```

```python
dataset = dataset.fillna(method='ffill')
```

```python
#Az X változó az összes attribútumot
#az y változó tartalmazza a cimket
```

```python
X = dataset[['Adult Mortality', 'infant deaths', 'Alcohol', 'Hepatitis B', 'GDP','Income composition of resources', 'Total expenditure', 'Schooling','Polio', 'Population','Income composition of resources', 'percentage expenditure']].values
y = dataset['Age'].values
```

```python
#varhato elettartam eloszlasa
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(dataset['Age'])
```

```python
#Ezután felosszuk az adatokat,  80% -át  az edzőkészletre, míg az adatok 20% -át a tesztkészletre.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Modell kepzese
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
#az alabi kod megjeleniti, hogy milyen együtthatókat választott a regressziós modell
```

```python
coeff_df = pd.DataFrame(regressor.coef_, ['Adult Mortality', 'infant deaths', 'Alcohol', 'Hepatitis B', 'GDP','Income composition of resources', 'Total expenditure', 'Schooling','Polio', 'Population','Income composition of resources', 'percentage expenditure'], columns=['Coefficient'])
coeff_df
```

```python
#előrejelzés a teszt adatokra
```

```python
y_pred = regressor.predict(X_test)
```

```python
#Ellenőrzi a tényleges és a becsült érték közötti különbséget.
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

```python
df1 = df.head(25)
df1
```

```python
#Most ábrázoljuk a tényleges és az előrejelzett értékek összehasonlítását
```

```python
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
